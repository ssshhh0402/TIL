```python
import random

class Pikachu:
    def __init__(self, name):
        self.name = name
        self.level= 5
        self.hp = self.level * random.randint(15, 20)
        self.exp = 0
        self.type = 'electric'
    def run(self):
        a = random.randint(0,1)
        return a
    
    def bark(self):
        print(f'{self.name}의 pikachu')
    
    def body_attack(self, someone):
        someone.hp -= self.level * 5
        print(f'{self.name}의 HP: {self.hp}')
        print(f'{someone.name}의 HP: {someone.hp}')
        if(someone.hp <= 0):
            print(f'{self.name}의 공격은 효과가 좋았다! {someone.name}은 쓰러졌다')
            self.exp += 15
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp = 0
    
    def thousond_volt(self, someone):
        someone.hp -= self.level * 7
        print(f'{self.name}의 HP: {self.hp}')
        print(f'{someone.name}의 HP: {someone.hp}')
        if(someone.hp <= 0):
            print(f'{self.name}의 공격은 효과가 좋았다! {someone.name}은 쓰러졌다')
            self.exp += 15
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp = 0

class Fire:
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * random.randint(15, 20)
        self.exp = 0
        self.type='fire'
   
    def run(self):
        a = random.randint(0,1)
        return a
    def bark(self):
        print(f'{self.name}의 Fire')
    
    def body_attack(self, someone):
        someone.hp -= self.level * 5
        print(f'{self.name}의 HP: {self.hp}')
        print(f'{someone.name}의 HP: {someone.hp}')
        if(someone.hp <= 0):
            print(f'{self.name}의 공격은 효과가 좋았다! {someone.name}은 쓰러졌다')
            self.exp += 15
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp = 0
    
    def thousond_volt(self, someone):
        if someone.type == 'green':
            someone.hp -= self.level * 10
        else:
            someone.hp -= self.level * 7
        print(f'{self.name}의 HP: {self.hp}')
        print(f'{someone.name}의 HP: {someone.hp}')
        if(someone.hp <= 0):
            print(f'{self.name}의 공격은 효과가 좋았다! {someone.name}은 쓰러졌다')
            self.exp += 15
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp = 0
            
class Water:
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * random.randint(15, 20)
        self.exp = 0
        self.type='water'
   
    def run(self):
        a = random.randint(0,1)
        return a
    def bark(self):
        print(f'{self.name}의 Fire')
    
    def body_attack(self, someone):
        someone.hp -= self.level * 5
        print(f'{self.name}의 HP: {self.hp}')
        print(f'{someone.name}의 HP: {someone.hp}')
        if(someone.hp <= 0):
            print(f'{self.name}의 공격은 효과가 좋았다! {someone.name}은 쓰러졌다')
            self.exp += 15
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp = 0
    
    def thousond_volt(self, someone):
        if someone.type == 'fire':
            someone.hp -= self.level * 10
        else:
            someone.hp -= self.level * 7
        print(f'{self.name}의 HP: {self.hp}')
        print(f'{someone.name}의 HP: {someone.hp}')
        if(someone.hp <= 0):
            print(f'{self.name}의 공격은 효과가 좋았다! {someone.name}은 쓰러졌다')
            self.exp += 15
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp = 0
            
class Green:
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * random.randint(15, 20)
        self.exp = 0
        self.type='green'
   
    def run(self):
        a = random.randint(0,1)
        return a
    def bark(self):
        print(f'{self.name}의 Fire')
    
    def body_attack(self, someone):
        someone.hp -= self.level * 5
        print(f'{self.name}의 HP: {self.hp}')
        print(f'{someone.name}의 HP: {someone.hp}')
        if(someone.hp <= 0):
            print(f'{self.name}의 공격은 효과가 좋았다! {someone.name}은 쓰러졌다')
            self.exp += 15
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp = 0
    
    def thousond_volt(self, someone):
        if someone.type == 'water':
            someone.hp -= self.level * 10
        else:
            someone.hp -= self.level * 7
        print(f'{self.name}의 HP: {self.hp}')
        print(f'{someone.name}의 HP: {someone.hp}')
        if(someone.hp <= 0):
            print(f'{self.name}의 공격은 효과가 좋았다! {someone.name}은 쓰러졌다')
            self.exp += 15
        if self.exp >= self.level * 100:
            self.level += 1
            self.exp = 0
```
